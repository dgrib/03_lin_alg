__3.__ Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>

$(x,y) = ||x||*||y||$
1) $(x,y)=(y,x) - выполняется;$<br>
2) $(\lambda x, y)=\lambda(x, y) - выполняется;$<br>
3) $(x_{1}+x_{2},y)=(x_{1},y)+(x_{2},y) - выполняется;$<br>

б) утроенное обычное скалярное произведение векторов?<br>

1) $3*(x,y)=3*(y,x) - выполняется;$<br>
2) $3*(\lambda x, y)=3*(\lambda(x, y)) - выполняется;$<br>
3) $3*(x_{1}+x_{2},y)=3*((x_{1},y)+(x_{2},y)) - выполняется;$<br>
